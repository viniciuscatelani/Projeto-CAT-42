# Cáluco da base complementar

## PMC:

### Quantidade do produtos: qTrib
### PMC do produto: PMC
### Percentual de desconto: perc_desconto
### vBCST retido da original: vBCST_ret_original

### vBCST = qTrib * (PMC * (1-(perc_desconto/100))
### vBCST_complementar = vBCST - vBCST_ret_original
### ICMS_ST = vBCST_complementar * Aliquota

## MVA:

### Valor do Produto: vProd
### MVA do produto: MVA
### vBCST = vProd * (1+(MVA/100))
### vBCST_complementar = vBCST
### ICMS_ST = vBCST_complementar * Aliquota

```python
novo_df_pmc['PMC_0'] = pd.to_numeric(novo_df_pmc['PMC_0'].str.replace(',', '.'), errors='coerce')
novo_df_pmc['PMC_12'] = pd.to_numeric(novo_df_pmc['PMC_12'].str.replace(',', '.'), errors='coerce')
novo_df_pmc['PMC_17'] = pd.to_numeric(novo_df_pmc['PMC_17'].str.replace(',', '.'), errors='coerce')
novo_df_pmc['PMC_17_5'] = pd.to_numeric(novo_df_pmc['PMC_17_5'].str.replace(',', '.'), errors='coerce')
novo_df_pmc['PMC_18'] = pd.to_numeric(novo_df_pmc['PMC_18'].str.replace(',', '.'), errors='coerce')
novo_df_pmc['PMC_20'] = pd.to_numeric(novo_df_pmc['PMC_20'].str.replace(',', '.'), errors='coerce')

conditions = [
    novo_df_pmc['ALIQUOTA'] == 0,
    novo_df_pmc['ALIQUOTA'] == 7,
    novo_df_pmc['ALIQUOTA'] == 12,
    novo_df_pmc['ALIQUOTA'] == 17,
    novo_df_pmc['ALIQUOTA'] == 17.5,
    novo_df_pmc['ALIQUOTA'] == 18,
    novo_df_pmc['ALIQUOTA'] == 20,
    novo_df_pmc['ALIQUOTA'] == 25,
]

values = [
    novo_df_pmc['PMC_0'],
    novo_df_pmc['PMC_12'],
    novo_df_pmc['PMC_12'],
    novo_df_pmc['PMC_17'],
    novo_df_pmc['PMC_17_5'],
    novo_df_pmc['PMC_18'],
    novo_df_pmc['PMC_20'],
    novo_df_pmc['PMC_20']
]

# Use np.select() to assign the appropriate value to valor_pmc based on conditions
valor_pmc = np.select(conditions, values, default=novo_df_pmc['ALIQUOTA']).astype(float)
novo_df_pmc['PMC'] = valor_pmc
# Calculate 'vbcst' column using vectorized arithmetic
novo_df_pmc['vBCST'] = novo_df_pmc['Quantidade Comercial'] * (novo_df_pmc['PMC'] * (1 - (novo_df_pmc['DESCONTO_PORCENTAGEM'] / 100)))
    
novo_df_pmc
```

```python
novo_df_pmc['CHAVE_ITEM'] = novo_df_pmc['CHAVE_ORIGINAL'].astype(str) + '-' + novo_df_pmc['COD_DO_ITEM'].astype(str)

# Filter out rows with NaN values in 'PMC' column
pmc_not_null = novo_df_pmc[novo_df_pmc['PMC'].notnull()]

# Use groupby and idxmax to efficiently get rows with the highest 'PMC' for each 'CHAVE_ITEM'
df_pmc_filt = pmc_not_null.loc[pmc_not_null.groupby('CHAVE_ITEM')['PMC'].idxmax()]

# Reset the index of the final DataFrame
df_pmc_filt.reset_index(drop=True, inplace=True)
df_pmc_filt
```

```python
df_pmc_null = novo_df_pmc[novo_df_pmc['PMC'].isnull()]
df_pmc_null['MVA_A_USAR'] = np.where((df_pmc_null['Data Emissão'] < df_pmc_null['DT_VALIDA']) | (df_pmc_null['MVA_DEPOIS'].isnull()),
                               df_pmc_null['MVA_ANTES'], df_pmc_null['MVA_DEPOIS'])

df_pmc_null
```

```python
df_pmc_novo = pd.concat([df_pmc_filt, df_pmc_null])
df_pmc_novo
```

```python
df_pmc_novo[df_pmc_novo['vBCST'].isnull()]
```

```python
df_pmc_novo['MVA_A_USAR'] = pd.to_numeric(df_pmc_novo['MVA_A_USAR'], errors='coerce')
df_pmc_novo['Valor Produto ou Serviço'] = df_pmc_novo['Valor Produto ou Serviço'].str.replace('.', '').str.replace(',', '.').astype(float)

def calculate_vBCST_final(row):
    if math.isnan(row['vBCST']):  # Check if vBCST is not null
        return row['Valor Produto ou Serviço'] * ((row['MVA_A_USAR'] / 100) + 1)
    else:
        return row['vBCST']

# Apply the function to create the new column 'vBCST_final'
df_pmc_novo['vBCST_final'] = df_pmc_novo.apply(calculate_vBCST_final, axis=1)

df_pmc_novo
```

```python
df_pmc_novo['vPMC'] = np.where(df_pmc_novo['PMC'].notnull(),
                              df_pmc_novo['PMC'],
                              df_pmc_novo['MVA_A_USAR'])

df_pmc_novo
```
