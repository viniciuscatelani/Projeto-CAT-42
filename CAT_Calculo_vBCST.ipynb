# Cáluco da base complementar

## PMC:

### Quantidade do produtos: qTrib
### PMC do produto: PMC
### Percentual de desconto: perc_desconto
### vBCST retido da original: vBCST_ret_original

### vBCST = qTrib * (PMC * (1-(perc_desconto/100))
### vBCST_complementar = vBCST - vBCST_ret_original
### ICMS_ST = vBCST_complementar * Aliquota

## MVA:

### Valor do Produto: vProd
### MVA do produto: MVA
### vBCST = vProd * (1+(MVA/100))
### vBCST_complementar = vBCST - vBCST_ret_original
### ICMS_ST = vBCST_complementar * Aliquota

```python
prods_mva['Valor Produto ou Serviço'] = prods_mva['Valor Produto ou Serviço'].str.replace('.', '').str.replace(',', '.').astype(float)
prods_mva
```
