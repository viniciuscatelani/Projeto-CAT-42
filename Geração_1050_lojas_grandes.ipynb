```python
# Importação das bibliotecas a serem utilizadas durante todo o processo

import pandas as pd
import numpy as np
from datetime import datetime
import math
```

```python
# Definição da loja a ser utilizada

loja = 28
```

```python
df_1 = pd.read_excel(f'Loja_{loja}/ficha_3_loja_{loja}_p1_v1.xlsb', sheet_name='ficha3', engine='pyxlsb')
df_1
```

```python
df_2 = pd.read_excel(f'Loja_{loja}/ficha_3_loja_{loja}_p2_v1.xlsb', sheet_name='ficha3', engine='pyxlsb')
df_2
```

```python
df = pd.concat([df_1, df_2])
df = df[['DATA','COD_ITEM', 'QTD_INI', 'ICMS_INI', 'SALDO_FINAL_MES_QTD', 'SALDO_FINAL_MES_ICMS']]
df
```

```python
df['REF'] = [datetime.strptime(x, '%Y-%m-%d').strftime('%Y%m') for x in df['DATA']]
df['REF'] = df['REF'].astype(str)
df
```

```python
grouped = df.groupby(['COD_ITEM', 'REF']).agg(
    QTD_INI=('QTD_INI', 'first'),
    ICMS_INI=('ICMS_INI', 'first'),
    SALDO_FINAL_MES_QTD=('SALDO_FINAL_MES_QTD', 'last'),
    SALDO_FINAL_MES_ICMS=('SALDO_FINAL_MES_ICMS', 'last')
).reset_index()

df_1050 = grouped.copy()
```

```python
df_1050
```

```python
cods_unicos = sorted(list(df['COD_ITEM'].unique()))
df['REF'] = [datetime.strptime(x, '%Y-%m-%d').strftime('%Y%m') for x in df['DATA']]
df['REF'] = df['REF'].astype(str)
refs_unicas = ['201910', '201911', '201912', '202001',
              '202002', '202003', '202004', '202005',
              '202006', '202007', '202008', '202009',
              '202010', '202011', '202012']
```

```python
df['REF'].unique()
```

```python
refs_unicas
```

```python
df_final = pd.DataFrame(columns=['REF', 'COD_ITEM'])
df_final['REF'] = sorted(refs_unicas*len(cods_unicos))
df_final['COD_ITEM'] = cods_unicos*len(refs_unicas)
df_final = df_final.sort_values(by=['COD_ITEM', 'REF']).drop_duplicates()
```

```python
df_final
```

```python
df_final['COD_ITEM'] = df_final['COD_ITEM'].astype(int)
df_1050['COD_ITEM'] = df_1050['COD_ITEM'].astype(int)
```

```python
merged = df_final.merge(df_1050,
                       left_on=['REF', 'COD_ITEM'], right_on=['REF', 'COD_ITEM'],
                       how='left')

merged = merged.sort_values(by=['COD_ITEM', 'REF'])
merged
```

```python
last_saldo_final_mes_qtd = None

for i in range(len(merged)):
    if pd.isnull(merged.at[i, 'QTD_INI']) and pd.isnull(merged.at[i, 'SALDO_FINAL_MES_QTD']):
        # Both QTD_INI and SALDO_FINAL_MES_QTD are null
        # Fill them with QTD_INI from the next row (i+1)
        if i < len(merged) - 1:
            merged.at[i, 'QTD_INI'] = merged.at[i+1, 'QTD_INI']
            merged.at[i, 'SALDO_FINAL_MES_QTD'] = merged.at[i+1, 'QTD_INI']
    
    elif pd.isnull(merged.at[i, 'SALDO_FINAL_MES_QTD']) and not pd.isnull(merged.at[i, 'QTD_INI']):
        # SALDO_FINAL_MES_QTD is null, but QTD_INI is not null
        # Fill SALDO_FINAL_MES_QTD with QTD_INI
        merged.at[i, 'SALDO_FINAL_MES_QTD'] = merged.at[i, 'QTD_INI']
    
    elif pd.isnull(merged.at[i, 'QTD_INI']) and not pd.isnull(merged.at[i, 'SALDO_FINAL_MES_QTD']):
        # QTD_INI is null, but SALDO_FINAL_MES_QTD is not null
        # Fill QTD_INI with SALDO_FINAL_MES_QTD from the previous row (i-1)
        if i > 0:
            merged.at[i, 'QTD_INI'] = merged.at[i-1, 'SALDO_FINAL_MES_QTD']
    
    # Check if both 'QTD_INI' and 'SALDO_FINAL_MES_QTD' are null from this point until the end
    if pd.isnull(merged.at[i, 'QTD_INI']) and pd.isnull(merged.at[i, 'SALDO_FINAL_MES_QTD']):
        # Fill them with the last known value of 'SALDO_FINAL_MES_QTD'
        merged.at[i, 'QTD_INI'] = last_saldo_final_mes_qtd
        merged.at[i, 'SALDO_FINAL_MES_QTD'] = last_saldo_final_mes_qtd
    else:
        # Update the last known value of 'SALDO_FINAL_MES_QTD'
        last_saldo_final_mes_qtd = merged.at[i, 'SALDO_FINAL_MES_QTD']
        
last_saldo_final_mes_icms = None

for i in range(len(merged)):
    if pd.isnull(merged.at[i, 'ICMS_INI']) and pd.isnull(merged.at[i, 'SALDO_FINAL_MES_ICMS']):
        # Both QTD_INI and SALDO_FINAL_MES_QTD are null
        # Fill them with QTD_INI from the next row (i+1)
        if i < len(merged) - 1:
            merged.at[i, 'ICMS_INI'] = merged.at[i+1, 'ICMS_INI']
            merged.at[i, 'SALDO_FINAL_MES_ICMS'] = merged.at[i+1, 'ICMS_INI']
    
    elif pd.isnull(merged.at[i, 'SALDO_FINAL_MES_ICMS']) and not pd.isnull(merged.at[i, 'ICMS_INI']):
        # SALDO_FINAL_MES_QTD is null, but QTD_INI is not null
        # Fill SALDO_FINAL_MES_QTD with QTD_INI
        merged.at[i, 'SALDO_FINAL_MES_ICMS'] = merged.at[i, 'ICMS_INI']
    
    elif pd.isnull(merged.at[i, 'ICMS_INI']) and not pd.isnull(merged.at[i, 'SALDO_FINAL_MES_ICMS']):
        # QTD_INI is null, but SALDO_FINAL_MES_QTD is not null
        # Fill QTD_INI with SALDO_FINAL_MES_QTD from the previous row (i-1)
        if i > 0:
            merged.at[i, 'ICMS_INI'] = merged.at[i-1, 'SALDO_FINAL_MES_ICMS']
    
    # Check if both 'QTD_INI' and 'SALDO_FINAL_MES_QTD' are null from this point until the end
    if pd.isnull(merged.at[i, 'ICMS_INI']) and pd.isnull(merged.at[i, 'SALDO_FINAL_MES_ICMS']):
        # Fill them with the last known value of 'SALDO_FINAL_MES_QTD'
        merged.at[i, 'ICMS_INI'] = last_saldo_final_mes_qtd
        merged.at[i, 'SALDO_FINAL_MES_ICMS'] = last_saldo_final_mes_qtd
    else:
        # Update the last known value of 'SALDO_FINAL_MES_QTD'
        last_saldo_final_mes_qtd = merged.at[i, 'SALDO_FINAL_MES_ICMS']
```
